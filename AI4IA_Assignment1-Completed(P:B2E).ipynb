```python
# B.9 part B:
from numpy import asarray
import numpy as np
import matplotlib.pyplot as plt
 
# objective function
def objective(x):
    return np.exp(-(x-3.14)**2)*(-1)
 
# derivative of objective function
def derivative(x):
    return np.exp(-(x-3.14)**2)*(2*x-6.28)

trajectory_x = []
trajectory_fx = []
 
# gradient descent algorithm
def gradient_descent(objective, derivative,n_iter,step_size):
  x= 3 # initial the first guess

  for i in range(n_iter):
        gradient = derivative(x) # calculate gradient
        x = x - step_size * gradient   # take a step
        fx = objective(x)  # evaluate candidate point
        # store solution
        trajectory_x .append(x)
        trajectory_fx.append(fx)
        # report progress
        print('>%d f(%s) = %.5f' % (i, x, fx))
  return [trajectory_x, trajectory_fx]
 


n_iter = 100  # define the total iterations
step_size = 0.1  # define the step size

# perform the gradient descent search
solutions, scores = gradient_descent(objective, derivative, n_iter, step_size)
# sample input range uniformly at 0.1 increments



# plotting
xx=np.linspace(0,6,1000)
plt.plot(xx, objective(xx), color='b',label='y=e^(-(x-3.14)^2)')
plt.plot(trajectory_x, trajectory_fx, color='r',label='particles')
plt.title(f'after {n_iter} iteration the final minimum place is in  f({trajectory_x[-1]})= {trajectory_fx[-1]}')
plt.legend(loc='lower left')
plt.xlabel("x ")
plt.ylabel("fx")
plt.show()
```

```python
# E.4 the core part 
import torch
from torch import autograd
import numpy as np
import math
import matplotlib.pyplot as plt


def f(x):
  return -1 * torch.exp(-torch.pow((x-np.pi),2)) + (0.01 * torch.pow((x-np.pi),2))

x=torch.tensor(-5.0,requires_grad=True) #initial guess
nr_steps = 1000

step_length1 = 1.0
step_length2 = 0.1
step_length3 = 0.01
```

```python
# E.4  with step_length1 = 1.0
traj_x=[]
traj_fx=[]
for i in range (nr_steps):
  fx=f(x)
  traj_x.append(x.item())
  traj_fx.append(fx.item())

  fx.backward()
  with torch.no_grad():
    x -= step_length1 * x.grad

  x.grad.zero_()
  #if some_stopping_criteria_fulfilled:
   # break

xx=torch.linspace(-5,11,requires_grad=True)
plt.plot(xx.detach().numpy(),f(xx).detach().numpy(),color='b',label="function")
plt.plot(traj_x,traj_fx,color='r',label="trajectory")
plt.legend(loc='lower left')
plt.xlabel("X ")
plt.ylabel("Y")
```

```python
# E.8

import torch
from torch import autograd
import math
import matplotlib.pyplot as plt

trajectory_t =[]
trajectory_c =[]


nr_steps = 10000
sp_len = 0.01

def C(t):
    return -60*(torch.exp(-0.1*t) - torch.exp(-0.3*t))
    
t = torch.tensor(10.0, requires_grad=True)

for i in range(nr_steps):
    ct = C(t)
    ct.backward()
    trajectory_t.append(t.detach().clone().numpy())
    trajectory_c.append(ct.item())
    with torch.no_grad():
        t -= sp_len * t.grad
    t.grad.zero_()

tt= torch.linspace(0,60,100)
plt.plot(tt, C(tt),label="function")
plt.plot(trajectory_t, trajectory_c, color = 'red', marker = '*',label="particles")
plt.title(f'after {nr_steps} of iteration c({t})= {C(t)} ')
plt.legend(loc='lower right')
plt.xlabel("t")
plt.ylabel("ct")
```
